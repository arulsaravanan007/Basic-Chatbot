```python
import nltk
import random
from nltk.chat.util import Chat, reflections
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import re
nltk.download('vader_lexicon')
sid = SentimentIntensityAnalyzer()
additional_patterns = [
    (r'weather', ['The weather is nice today!', 'It seems like a sunny day!', 'I am not sure about the weather, sorry.']),
    (r'hobby', ['My hobby is chatting with you!', 'I enjoy learning new things.', 'I like to read and learn.']),
    (r'favorite food', ['I do not eat, but I hear biryani is quite popular!', 'I do not have taste buds, but I have heard good things about butter chicken.']),
    (r'i am (.*)', ['That sounds wonderful!', 'Glad to hear that!', 'Great to know!']),
    (r'features|capabilities|work', ["I can chat with you on various topics.", "I'm capable of providing information on weather, hobbies, favorite foods, and more.", "I'm designed to engage in conversation and assist you with your inquiries."]),
    (r'hello|hey', ['Hello!', 'Hi there!', 'Hey!']),
    (r'how are you', [ 'I am doing well, thank you for asking.']),
    (r'goodbye|bye', ['Goodbye!', 'See you later!', 'Have a great day!']),
    (r'thank you', ['You are welcome!', 'Glad I could help!', 'Anytime!']),
]
patterns = [
    (r'hi', ['Hello!,How are you today?']),
]
patterns += additional_patterns
class AdvancedChatbot:
    def __init__(self):
        self.chatbot = Chat(patterns, reflections)
    def start(self):
        print("Hello! I'm a friendly chatbot. How can I assist you today?")
        self.chat()
    def analyze_sentiment(self, text):
        score = sid.polarity_scores(text)
        if score['compound'] >= 0.05:
            return 'positive'
        elif score['compound'] <= -0.05:
            return 'negative'
        else:
            return 'neutral'
    def get_sentiment_pattern(self, sentiment):
        sentiment_patterns = {
            'positive': ['That sounds wonderful!', 'Glad to hear that!', 'Great to know!'],
            'negative': ['I\'m sorry to hear that.', 'It\'s alright, things will get better.', 'I understand, life can be tough sometimes.'],
        }
        return random.choice(sentiment_patterns[sentiment])
    def check_additional_patterns(self, user_input):
        for pattern, responses in additional_patterns:
            if re.search(pattern, user_input):
                return random.choice(responses)
        return None

    def chat(self):
        while True:
            user_input = input("You: ")
            if user_input.lower() == 'quit':
                print("Goodbye!")
                break
            else:
                sentiment = self.analyze_sentiment(user_input)
                if sentiment in ['positive', 'negative']:
                    pattern_response = self.get_sentiment_pattern(sentiment)
                    print("Bot:", pattern_response)
                else:
                    additional_response = self.check_additional_patterns(user_input)
                    if additional_response:
                        print("Bot:", additional_response)
                    else:
                        pattern_response = self.chatbot.respond(user_input)
                        print("Bot:", pattern_response)

def main():
    chatbot = AdvancedChatbot()
    chatbot.start()
if __name__ == "__main__":
    main()
```
